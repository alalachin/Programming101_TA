```python
# 看得懂了！ 再丟一次進去pyquery
doc = pq(res.text)
doc(".content .site-list").text()
```

```python
#做迭代，印出每個大類別（元素）的網址，存在site-list這個類別，它裡面就是a標籤 

for eachLV1doc in doc(".content .site-list").items():
    print (eachLV1doc("a").attr("href"))
    
#但是印出來是相對路徑... -->針對doc做調整！
```

```python
# 印出絕對路徑
doc.make_links_absolute(base_url = res.url)
for eachLV1doc in doc(".content .site-list").items():
    print (eachLV1doc("a").attr("href"))
# 可以取得大類別的連結了
```

```python
# 把連結丟到pyquery裡頭，是可以解析的

# 點開大類別，開始處理中類別
# 左邊最上面東西（促銷）也會出現在底下，可以排除
# 觀察中類別的selector, 發現是一個不能用的ID，因為都是數字，會變！
# ID不應該做為找東西的終點，而是起點
# 所以要找一個不會變、夠特別、包含所有中類別的ID --> 一直往上找！
# 得"cl-menucate"

for eachLV1doc in doc(".content .site-list").items():
    LV1doc = pq(eachLV1doc("a").attr("href"))
    print(LV1doc("#cl-menucate .sitelist .list").text())
# 把所有的中類別列出來 (太多了，要按停止)
```

```python
# 接著把促銷區的項目去除掉( 印出來的都是中類別)
for eachLV1doc in doc(".content .site-list").items():
    LV1doc = pq(eachLV1doc("a").attr("href"))
    print(LV1doc("#cl-menucate .sitelist:nth-child(n+2) .list").text())
    # print(LV1doc("#cl-menucate .sitelist .list").text())
```

```python
# 找到“促銷區”字樣
# 要cl-menucate裡面的.sitelist的第一個孩子裡面的.stitle類別 
for eachLV1doc in doc(".content .site-list").items():
    LV1doc = pq(eachLV1doc("a").attr("href"))
    print(LV1doc("#cl-menucate .sitelist:nth-child(1) .stitle").text())
```

```python
# 去除這些字 
for eachLV1doc in doc(".content .site-list").items():
    LV1doc = pq(eachLV1doc("a").attr("href"))
    if "活動" in LV1doc("#cl-menucate .sitelist:nth-child(1) .stitle").text():
        print (LV1doc("#cl-menucate .sitelist:nth-child(n+2) .list").text())
        print (1)
    else:
        print(LV1doc("#cl-menucate .sitelist .list").text())
        print (2)
```

```python
# 最好都寫成函示 （老師把這邊寫在前面import下面）沒有要寫新的東西，把原本的東西丟進來而已
# 老師說寫爬蟲最好控制在一個for迴圈

# index_page幫我們把首頁打開、把大類別的連結找出來
def index_page(index_url):
    res = requests.get(index_url)
    res.encoding="big5"
    doc = pq(res.text)               # 把網頁的內容用pyquery存到doc中
    doc.make_links_absolute(base_url = res.url)
    for eachLV1doc in doc(".content .site-list").items():
        # LV1doc = pq(eachLV1doc("a").attr("href"))
        LV1_page(eachLV1doc("a").attr("href"))
#         if "活動" in LV1doc("#cl-menucate .sitelist:nth-child(1) .stitle").text():
#             print (LV1doc("#cl-menucate .sitelist:nth-child(n+2) .list").text())
#         else:
#             print(LV1doc("#cl-menucate .sitelist .list").text())
```

```python
# 把每個連結打開、並判斷裡面的中類別 
def LV1_page(url):
    LV1doc = pq(url)
    LV1doc.make_links_absolute()
    if "活動" in LV1doc("#cl-menucate .sitelist:nth-child(1) .stitle").text():
        for eachLV2doc in LV1doc("#cl-menucate .sitelist:nth-child(n+2) .list").items():
            LV2_page(eachLV2doc("a").attr("href"))
    else:
        for eachLV2doc in LV1doc("#cl-menucate .sitelist .list").items():
            LV2_page (eachLV2doc("a").attr("href"))
```
